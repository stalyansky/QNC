```python
#Written by Seth Talyansky on September 9, 2025
import numpy as np
from scipy import stats
mu = 10
s = 2
ns = [5, 10, 20, 40, 80, 160, 1000]

for n in ns:
    print('n = %d'%n)
    sample = np.random.normal(mu, s, n)
    #Method 1
    sem = s/n**0.5
    ci = [np.round(mu-1.96*sem, 2), np.round(mu+1.96*sem, 2)]
    print('First confidence interval is %f, %f'%(ci[0], ci[1]))
    
    #Method 2
    sem = s/n**0.5
    ci = [np.round(mu+stats.t.ppf(0.025, n)*sem, 2), np.round(mu-stats.t.ppf(0.025, n)*sem, 2)]
    print('Second confidence interval is %f, %f'%(ci[0], ci[1]))
    
    #Method 3
    means = []
    for i in range(1000):
        means.append(np.mean(np.random.choice(sample, size=n)))
    ci = [np.round(np.quantile(means, 0.025), 2), np.round(np.quantile(means, 0.975), 2)]
    print('Third confidence interval is %f, %f'%(ci[0], ci[1]))
    
    #Method 4
    ci = np.array(stats.bayes_mvs(sample, alpha=0.95)[0][1])
    ci[0] = np.round(ci[0], 2)
    ci[1] = np.round(ci[1], 2)
    print('Fourth confidence interval is %f, %f'%(ci[0], ci[1]))
    print('\n')
```
