```python
#Written by Seth Talyansky on September 19, 2025
import numpy as np
from scipy import stats
#Exercise parts 1 and 2
n = 10
n_trials = 1000
alpha = 0.05
bonferroni_alpha = alpha/n_trials
p_values = []
for i in range(n_trials):
    sample1 = stats.norm.rvs(size=n)
    sample2 = stats.norm.rvs(size=n)
    result = stats.ttest_ind(sample1, sample2)
    p_values.append(result.pvalue)
p_values = np.sort(np.array(p_values))
critical_values = []
for i in range(1, n_trials+1):
    critical_values.append(i/n_trials*alpha)
critical_values = np.array(critical_values)
c = 0
diff = critical_values-p_values
greater = np.where(diff > 0)[0]
if len(greater > 0):
    c = max(greater) + 1
bh_passed = np.round(c/n_trials*100, 2)
bonferroni_passed = np.round(100*sum(p_values < bonferroni_alpha)/n_trials, 2)
print('Part 2: samples taken from the same distribution with mean 0')
print('The percentage of P-values significant after Bonferroni correction is %d%%'%bonferroni_passed)
print('The percentage of P-values significant after BH correction is %d%%'%bh_passed)

#Exercise part 3
n = 10
n_trials = 1000
alpha = 0.05
bonferroni_alpha = alpha/n_trials
p_values = []
for i in range(n_trials):
    sample1 = stats.norm.rvs(loc=1, size=n)
    sample2 = stats.norm.rvs(loc=2, size=n)
    result = stats.ttest_ind(sample1, sample2)
    p_values.append(result.pvalue)
p_values = np.sort(np.array(p_values))
critical_values = []
for i in range(1, n_trials+1):
    critical_values.append(i/n_trials*alpha)
critical_values = np.array(critical_values)
c = 0
diff = critical_values-p_values
greater = np.where(diff > 0)[0]
if len(greater > 0):
    c = max(greater) + 1
bh_passed = np.round(c/n_trials*100, 2)
bonferroni_passed = np.round(100*sum(p_values < bonferroni_alpha)/n_trials, 2)
print('Part 3: samples taken from different distributions with mean 1 and mean 2')
print('The percentage of P-values significant after Bonferroni correction is %d%%'%bonferroni_passed)
print('The percentage of P-values significant after BH correction is %d%%'%bh_passed)
p_values = []
for i in range(n_trials):
    sample1 = stats.norm.rvs(loc=1, size=n)
    sample2 = stats.norm.rvs(loc=4, size=n)
    result = stats.ttest_ind(sample1, sample2)
    p_values.append(result.pvalue)
p_values = np.sort(np.array(p_values))
critical_values = []
for i in range(1, n_trials+1):
    critical_values.append(i/n_trials*alpha)
critical_values = np.array(critical_values)
c = 0
diff = critical_values-p_values
greater = np.where(diff > 0)[0]
if len(greater > 0):
    c = max(greater) + 1
bh_passed = np.round(c/n_trials*100, 2)
bonferroni_passed = np.round(100*sum(p_values < bonferroni_alpha)/n_trials, 2)
print('Part 3a: samples taken from different distributions with mean 1 and mean 4')
print('The percentage of P-values significant after Bonferroni correction is %d%%'%bonferroni_passed)
print('The percentage of P-values significant after BH correction is %d%%'%bh_passed)
print('As the difference between the means of the sample distributions increases, t-tests on random samples are more likely to yield results that survive both types of multiple comparison correction')
```
