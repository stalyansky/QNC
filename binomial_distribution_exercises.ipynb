```python
#Written by Seth Talyansky on September 10, 2025
import numpy as np
from scipy import stats

#Exercise 1
print('Exercise 1:')
p = 0.2
n = 10
for k in range(1, 11):
    pk = stats.binom.pmf(k, n, 0.2)
    print('The probability that %d quantum/a will be released is %f'%(k, pk))

#Exercise 2
print('\nExercise 2:')
probabilities = np.arange(0.1, 1.1, 0.1)
likelihoods8 = []
for p in probabilities:
    l = stats.binom.pmf(8, 14, p)
    likelihoods8.append(l)
    print('The likelihood that 8 of 14 quanta were released given a release probability of %f is %f'%(p, l))
likelihoods8 = np.array(likelihoods8)
ml = probabilities[np.where(likelihoods8 == np.max(likelihoods8))[0][0]]
print('The likeliest release probability is %f'%ml)

#Exercise 3
print('\nExercise 3:')
probabilities = np.arange(0.1, 1.1, 0.1)
likelihoods5 = []
for p in probabilities:
    l = stats.binom.pmf(5, 14, p)
    likelihoods5.append(l)
likelihoods5 = np.array(likelihoods5)
likelihoods_total = likelihoods8*likelihoods5
likelihoods_total_log = np.log(likelihoods_total)
ml = probabilities[np.where(likelihoods_total == np.max(likelihoods_total))[0][0]]
for i in range(len(probabilities)):
    print('Given a release probability of %f, the total likelihood is %f and the total log-likelihood is %f'%(probabilities[i], likelihoods_total[i], likelihoods_total_log[i]))
print('The likeliest release probability is %f'%ml)

probabilities = np.arange(0.01, 1.01, 0.01)
likelihoods_total = []
for p in probabilities:
    l = stats.binom.pmf(8, 14, p)*stats.binom.pmf(5, 14, p)
    likelihoods_total.append(l)
likelihoods_total = np.array(likelihoods_total)
ml = probabilities[np.where(likelihoods_total == np.max(likelihoods_total))[0][0]]
print('The likeliest release probability using a resolution of 0.01 is %f'%ml)
probabilities = np.arange(0.001, 1.001, 0.001)
likelihoods_total = []
for p in probabilities:
    l = stats.binom.pmf(8, 14, p)*stats.binom.pmf(5, 14, p)
    likelihoods_total.append(l)
likelihoods_total = np.array(likelihoods_total)
ml = probabilities[np.where(likelihoods_total == np.max(likelihoods_total))[0][0]]
print('The likeliest release probability using a resolution of 0.001 is %f'%ml)
print('The estimate of true release probability becomes more precise as resolution increases')

#Exercise 4
print('\nExercise 4:')
measurements = [2]*3 + [3]*7 + [4]*10 + [5]*19 + [6]*26 + [7]*16 + [8]*16 + [9]*5 + [10]*5
probabilities = np.arange(0.01, 1.01, 0.01)
likelihoods_total = np.ones(len(probabilities))
for m in measurements:
    likelihoods = []
    for p in probabilities:
        likelihoods.append(stats.binom.pmf(m, 14, p))
    likelihoods_total = likelihoods_total*np.array(likelihoods)
ml = probabilities[np.where(likelihoods_total == np.max(likelihoods_total))[0][0]]
print('The likeliest release probability (p hat) is %f'%ml)

#Exercise 5
print('\nExercise 5:')
probabilities = np.arange(0.01, 1.01, 0.01)
likelihoods = []
for p in probabilities:
    l = stats.binom.pmf(7, 14, p)*stats.binom.pmf(7, 14, p)
    likelihoods.append(l)
likelihoods = np.array(likelihoods)
ml = probabilities[np.where(likelihoods == np.max(likelihoods))[0][0]]
print('The likeliest release probability (p hat) given 7/14 successes  is %f'%ml)
pc = 1-stats.binom.cdf(6, 14, 0.3)
print('The probability of 7 released quanta or more out of 14 under the null hypothesis is %f'%pc)
print('The null hypothesis thus cannot be rejected at the alpha = 0.05 confidence level because P > 0.05, so it cannot be concluded that temperature had an effect')
```
