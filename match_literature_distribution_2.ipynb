```python
#Written by Seth Talyansky on September 7, 2025
#Code adapted from binomial distribution tutorial to replicate Fig. 2A/C from Searl and Silinsky (J. Physiol. 2002)

import numpy as np
import collections
import matplotlib.pyplot as plt
from scipy.stats import poisson

m = 2.1                    
num_experiments = 10000;     # number of "experiments"
outcomes = poisson.rvs(m,size=num_experiments)

counts = collections.Counter(outcomes)
plt.bar(counts.keys(), counts.values())
plt.title(f'Calcium-mediated MEPPs 100 1/ms (Poisson model), lambda={m}, {num_experiments} simulations')
plt.ylabel('Count')
plt.show()
```
